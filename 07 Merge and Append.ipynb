```python
cost = pd.DataFrame({
    "NAME" : ["A","B","C"],
    "CP"   : [50,100,200]
})
```

# SIDE BY SIDE MERGE

```python
# 'pd.concat' function is used to combine multiple DataFrames or Series.
# Yaha hum 'cost' aur 'sell' DataFrame ko combine kar rahe hain.

# axis=1 ka matlab hai join columns side by side (horizontal concatenation).
# Matlab dono DataFrames ko ek saath columns ke roop me add karega.

pd.concat([cost, sell], axis=1)  # Yeh dono DataFrames ko combine kar ke ek naya DataFrame bana dega [web:1][web:5][web:12]
```

```python
# cost aur sell DataFrames ko merge kar rahe hain
# on parameter ke through 'NAME' column ke basis par join kiya ja raha hai
# outer ka matlab hai ki dono DataFrames ke sabhi rows include hongi,
# agar koi row sirf ek DataFrame mein hai to uski jagah NaN aa jayega

pd.merge(cost, sell, on = ['NAME','NAME'], how = 'outer')

# explaination:
# merge() function dataframes ko join karta hai, commonly database jaise join operations ke tarah
# isme 'on' parameter specify karta hai ki kis column ke basis par join karna hai
# 'how' parameter se join ki type decide hoti hai, yahan 'outer' hai jiska matlab hai full outer join
```

```python
pd.merge(cost , sell, on = ['NAME','NAME'], how = 'inner')
```

```python
pd.merge(cost , sell, on = ['NAME','NAME'], how = 'left')
```

```python
pd.merge(cost , sell, on = ['NAME','NAME'], how = 'right')
```

```python
# cost aur sell DataFrames ko merge karte hain
# cost DataFrame ka 'CP' column aur sell DataFrame ka 'SP' column match karenge
# left_on = 'CP' matlab cost DataFrame ka CP column use hoga
# right_on = 'SP' matlab sell DataFrame ka SP column use hoga
# yeh merge operation dono DataFrames ko combine karega jahan CP aur SP values match karenge
pd.merge(cost, sell, left_on=['CP'], right_on='SP')
```
