# loc and iloc

```python
d = {
    'name': ['sahil', 'abhi', 'harjeet'],
    'age': [20, 23, 21]
}
```

```python
df = pd.DataFrame(d)
```

```python
df
```

```python
df = pd.DataFrame(d, index=['A', 'B', 'C'])
```

```python
df
```

#### loc and iloc = The main difference between loc and iloc in pandas is that loc is label-based indexing (using row and column names),
#### while iloc is integer position-based indexing (using row and column numbers)

### filtering by index name

```python
df.loc['A']
```

## filtering by index number

```python
df.iloc[1]
```

```python
a= df[0:1]
```

```python
a
```

```python
a.iloc[0,0]
```

```python
df.iloc[1:,0]
```
